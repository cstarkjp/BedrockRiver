## Python setup

```python
import bedrockriver as br, sympy as sy, numpy as np, matplotlib.pyplot as plt
sy.init_printing(pretty_print=True,wrap_line=True) #,num_columns=20 IPython ignores options
from symbols import *
```

```python
sm = br.sinuosity.bend_model()
```

```python
sm.get_params()
```

## Solutions

```python
# ss = sinuosity.set(sm=sm, Q_range=(500,3000), Q_step=250)
sm.reset_params()
ss = br.ensemble.set(sm=sm, Q_range=(500,3000), Q_step=500)
sm.set_params(  {beta_0:0.0015, d_i:50, mu:1} )
sm.get_params()
ss.initial_w_range_array
```

```python
ss.solve_set_w_chi(n_solutions=2, t_span=(0.0,1e3),
                   initial_chi_range=(0.001,0.1) )
```

```python
br.plot.ensemble_chi_w_for_t(sm,ss)
```

```python
br.plot.ensemble_mean_chi_for_Q(sm,ss)
```

```python
br.plot.ensemble_mean_w_for_Q(sm,ss)
```
