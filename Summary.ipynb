## Why does the sinuosity of incising rivers positively correlate with "storminess" and relative flood intensity?

  - shown by Stark et al, 2010
  - for mixed bedrock-alluvial channels in island mountains around the western North Pacific
  - "regional" mean sinuosity measured by averaging reach sinuosities across similar-size catchments
  - for PNG/Philippines/Taiwan/Japan, we found sinuosity to correlate with:
      - relative flood intensity (coefficient of variation of flood discharge)
      - relative rainfall magnitude
      - typhoon strike frequency
  - for Japanese rivers, the correlation was shown to be greater on more erodible bedrock, less pronounced on less erodible bedrock => _sinuosity correlates with bedrock erodibility_
  - satellite imaging confirmed that stronger sinuosity is equivalent to a greater degree of channel meandering =>  _sinuosity ~ degree of meander development_

## Why should regionally averaged sinuosity correlate with anything?


  - Meander models/theory indicate bend migration continues until cutoff (loop or chute)
      - no other morphodynamic mechanism exists to stop meander growth
      - given time, all mountain rivers should therefore meander until such cutoff
      - average sinuosity should ultimately reach the same value everywhere
      - sinuosity across regions should not correlate with anything
  - /Either/ an _external resetting mechanism_ must exist whose reset frequency correlates with storminess
     - candidate reset mechanisms include:
         - propagation of a big knickpoint
         - massive landsliding
         - overwhelming valley alluviation
         - climage changes big enough to drastically alter flood discharge => bend wavelength
         - drainage capture changing upstream area => flood discharge => bend wavelength
  - /Or/ some _slowdown mechanism_ must exist:
     - meander growth rate & slowdown time scales must be functions of "storminess"


## Model goals, constraints and ideas


1. Build a model that can explore how bedrock channel sinuosity evolves
   - locally: at the bend scale
   - regionally: at the catchment scale per Stark et al 2010
2. Make it as parsimonious as possible
   - initially omitting e.g. role of bed cover?
   - simplify meander morphodynamics & channel flow as drastically as possible
3. Consider bend apex only
   - and use fixed wavelength
4. Develop equations into a dynamical system with 2DOF or 3DOF
   - system of ODEs in $t$ for simplified channel cross-section at bend only
   - 2DOF: channel width $w(t)$ and sinuosity $\chi(t)$ at fixed bend-bend height drop $S_{ref}$
   - 3DOF: channel width $w(t)$, sinuosity $\chi(t)$ and bend-bend slope $S_ref(t)$
5. Use ensemble of independent model bend evolutions to mimic regional behavior
   - employ trick of Stark & Passalacqua 2014
   - each bend evolves in parallel over randomly varied durations
   - _regional average_ equates to _time-average_ of all points along traces on phase portrait

## Work so far on *Sinuosity* model

1.  Bedrock channel cross-section is simplified into a symmetric (isoceles) trapezoid with wall angle $\theta$:
  - imbalance between rates of wall-normal erosion $\xi_y$ & vertical bed erosion $\xi_z$ drives channel widening vs narrowing
  - if wall erosion rate is zero, bed erosion drives half-channel narrowing
  - if bed erosion rate is zero, wall erosion drives half-channel widening
  - imbalance in left vs right wall erosion rates drives channel centerline motion

2. The motions of the walls and the bed can be expressed as left and right "corner" vectors
  - The combined corner vectors define the rates of channel widening and lateral channel motion
  - At fixed wavelength, lateral channel motion can be written as rate of sinuosity change
